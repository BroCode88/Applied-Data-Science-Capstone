```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Model': ['Logistic Regression', 'SVM', 'Decision Tree', 'KNN'],
    'Score': [0.846429, 0.848214, 0.887500, 0.848214]
}
df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(df['Model'], df['Score'], color=['orange', 'green', 'blue', 'red'])  # Create the bar chart


plt.title('Model Performance Comparison', fontsize=16)  # Title of the chart
plt.xlabel('Models', fontsize=14)  # X-axis label
plt.ylabel('Scores', fontsize=14)  # Y-axis label
plt.ylim(0.8, 1)  # Set the limit for the Y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better readability


plt.show()
```
