```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

df = pd.read_csv(url)

success_rate = df.groupby('Orbit')['Class'].mean().reset_index()
success_rate.columns = ['Orbit', 'Success Rate']

df = df.merge(success_rate, on='Orbit', how='left')

plt.figure(figsize=(14, 8))
sns.scatterplot(x="Success Rate", y="Orbit", hue="Class", data=df, palette="deep", alpha=0.7)

plt.xlabel("Success Rate", fontsize=20)
plt.ylabel("Orbit", fontsize=20)
plt.title("Scatter Plot of Success Rate by Orbit Type", fontsize=24)
plt.legend(title='Class', loc='upper right')
plt.show()
```
